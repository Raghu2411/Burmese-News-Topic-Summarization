```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset into a pandas DataFrame
df = pd.read_csv('sum_data_languages.csv')

# Assuming the dataset has columns Language, Train, Dev, Test, Total
# If column names differ, adjust accordingly
languages = df['Language']
train = df['Train']
dev = df['Dev']
test = df['Test']
total = df['Total']

x = range(len(languages))

plt.figure(figsize=(10, 6))

plt.bar(x, train, width=0.2, label='Train')
plt.bar(x, dev, width=0.2, label='Dev', bottom=train)
plt.bar(x, test, width=0.2, label='Test', bottom=train + dev)
plt.bar(x, total, width=0.2, label='Total', bottom=train + dev + test)

plt.xlabel('Languages')
plt.ylabel('Count')
plt.title('Distribution of Train, Dev, Test, Total with respect to Language')
plt.xticks(x, languages)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset into a pandas DataFrame
df = pd.read_csv('sum_data_languages.csv')

# Find the top 9 languages with the highest total amount
top_10_languages = df.nlargest(10, 'Total')['Language'].tolist()

# Include Burmese
selected_languages = top_10_languages + ['Burmese']+ ['Japanese']+ ['Chinese (Traditional)']+ ['Chinese (Simplified)']+ ['Korean']+ ['Tamil']

# Filter the DataFrame to include only the selected languages
selected_df = df[df['Language'].isin(selected_languages)]

# Plotting for Total
plt.figure(figsize=(10, 6))
bars = plt.bar(selected_df['Language'], selected_df['Total'])
plt.xlabel('Languages')
plt.ylabel('Total Count')
plt.title('Distribution of Total with respect to Language')
plt.xticks(rotation=45)
plt.tight_layout()

# Adding total count on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 50, round(yval), ha='center', color='black', fontsize=9)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset into a pandas DataFrame
df = pd.read_csv('sum_data_languages.csv')

# Sort the DataFrame by the 'Total' column in descending order and select the top 10 rows
top_10_languages = df.sort_values(by='Total', ascending=False).head(10)

# Include Burmese
top_10_languages = top_10_languages.append(df[df['Language'] == 'Burmese'])

languages = top_10_languages['Language']
train = top_10_languages['Train']
dev = top_10_languages['Dev']
test = top_10_languages['Test']
total = top_10_languages['Total']

x = range(len(languages))

plt.figure(figsize=(10, 6))

plt.bar(x, train, width=0.2, label='Train')
plt.bar(x, dev, width=0.2, label='Dev', bottom=train)
plt.bar(x, test, width=0.2, label='Test', bottom=train + dev)
plt.bar(x, total, width=0.2, label='Total', bottom=train + dev + test)

plt.xlabel('Languages')
plt.ylabel('Count')
plt.title('Distribution of Train, Dev, Test, Total with respect to Language (Top 10)')
plt.xticks(x, languages)
plt.legend()

plt.tight_layout()
plt.show()
```
